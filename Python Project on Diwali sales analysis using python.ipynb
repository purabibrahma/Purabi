

Product Category



```python
sns.set(rc={'figure.figsize':(25,8)})
ax=sns.countplot(data=df, x='Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#To see Cateogry wise sales in terms of orders

sales_occupation=df.groupby(['Product_Category'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)

sns.set(rc={'figure.figsize':(20,8)})
sns.barplot(data=sales_occupation, x='Product_Category', y='Orders')
```

```python
##To see Cateogry wise sales in terms of amount


sales_occupation=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,8)})
sns.barplot(data=sales_occupation, x='Product_Category', y='Amount')
```

```python
##To see to 10 Cateogries of sales in terms of amount

sales_occupation=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,8)})
sns.barplot(data=sales_occupation, x='Product_Category', y='Amount')
```

From the above graph, we can see that higest selling category in terms or orders are Clothing & Apperal, Food and electronics  & Gadgets respectively.

But in terms of amount contribution, top categories are Food, Clothing & Apperal and electronics and Gudgets respecticely.

```python
##To see to 10 Product_ID giving maximum sales

sales_productid=df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,8)})
sns.barplot(data=sales_productid, x='Product_ID', y='Orders')
```

```python
#top 10 most sold products (Same thing as above)

fig1, ax1 = plt.subplots(figsize=(12,7))

df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
